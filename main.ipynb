```python
# function to extract values of name
import ast
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):   # ast.literal_eval:   it'll become list and able to get the values of name
        lst.append(i['name'])
    return lst
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
# function to extract values of name
import ast
def convert2(obj):
    lst = []
    for i in ast.literal_eval(obj):
        counter =0
        if counter !=3:
               # ast.literal_eval:   it'll become list and able to get the values of name
            lst.append(i['name'])
            counter+=1
        else:
            break
    return lst
```

```python
movie['cast'] = movie['cast'].apply(convert2)
```

```python
def get_directors(obj):
    lst = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            lst.append(i['name'])
            break
    return lst
```

```python
movie['crew']= movie['crew'].apply(get_directors)
```

```python
# Convert each string in the 'overview' column to a list of words
movie['overview']= movie['overview'].apply(lambda x: x.split() if isinstance(x, str) else x)
```

```python
movie.head()
```

```python
# removing spaces between words
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
# concatination of tag columns
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords']+ movie['cast'] + movie['crew']
```

```python
# creating new data frame
movie_2 = movie[["id","title","tags"]]
```

```python
# converting into string
movie_2['tags'] = movie_2['tags'].apply(lambda x:" ".join(x))
```

```python
movie_2['tags'] = movie_2['tags'].apply(lambda x:x.lower())
```

```python
movie_2.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps  = PorterStemmer()

def stem(text):
    result = []
    for i in text.split():
        result.append(ps.stem(i))
    
    return " ".join(result)

movie_2['tags']= movie_2['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
# max_featues are the words used frequently, we need to remove or ignore stop words of english lagnuage
```

```python
# convert words into array/ vector
vectors = cv.fit_transform(movie_2['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x:x[1])[1:11]
```

```python
def recommend(mov):
    movie_index = movie_2[movie_2['title']== mov].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True, key= lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(movie_2.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
